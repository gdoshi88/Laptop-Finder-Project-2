```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
both_csv = 'both_laptops.csv'
```

```python
both_data_df = pd.read_csv(both_csv, '|', header=0)
# both_data_df.head()
both_data_df = both_data_df[['brand', 'model', 'screenSize', 'cpu', 'ram', 'hd', 'upc', 'price', 'link', 'title', 'lastUpdated']]
both_data_df.head()    

both_data_df.rename(columns={
    'screenSize': 'screensize'
}, inplace = True)
both_data_df.head()
```

```python
###############################################################
# LOAD: 
###############################################################


# change to your username:password@localhost:5432/laptop_db:
connection = "postgres:YOURPASSWORDHERE@localhost:5432/laptop_db"
engine = create_engine(f'postgresql://{connection}')


# add the BOTH laptop dataframes to the table:
both_data_df.to_sql(name="both_laptops", con=engine, if_exists='replace', index=False)


# VALIDATE TABLE BUILD:
engine.table_names()
```
