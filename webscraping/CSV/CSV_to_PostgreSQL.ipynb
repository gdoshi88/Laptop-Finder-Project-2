```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
bb_csv = 'bb_laptops.csv'
frys_csv = 'frys_laptops.csv'
```

```python
bb_data_df = pd.read_csv(bb_csv, '|', header=None)
bb_data_df.head()


bb_data_df.rename(columns={
    0: 'brand',
    1: 'model',
    2: 'screensize',
    3: 'cpu',
    4: 'ram',
    5: 'hd',
    6: 'graphics',
    7: 'upc',
    8: 'price',
    9: 'link', 
    10: 'title',
    11: 'lastUpdated'
}, inplace = True)
bb_data_df.head()
```
